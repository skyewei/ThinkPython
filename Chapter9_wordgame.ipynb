```python
## 9.6 利用for循环

def is_abecedarian(word):
    previous = word[0]
    for c in word:
        if c < previous:
            return False
        previous = c
    return True

def main():
    count = 0
    for line in open('words.txt'):
        if is_abecedarian(line.strip()):
            count += 1
    print(count)
    
main()
```

```python
## 采用递归

def is_abecedarian(word):
    if len(word) <= 1:
        return True
    if word[0] > word[1]:
        return False
    return is_abecedarian(word[1:])

def main():
    count = 0
    for line in open('words.txt'):
        if is_abecedarian(line.strip()):
            count += 1
    print(count)
        
main()
```

```python
## 9.7 找到一对连续双字符

def find_one_double(word):
    if len(word) < 2:
        return False
    if word[0] == word[1]:
        return True
    return find_one_double(word[1:])

def main():
    one_double = []
    for line in open('words.txt'):
        if find_one_double(line.strip()):
            one_double.append(line.strip())
    print(len(one_double))
    
main()
```

```python
## 9.7 找到三个连续双字符

def find_three_double(word, count=0):
    if count < 3:
        if len(word) < 2:
            return False
        if word[0] == word[1]:
            count += 1
            return find_three_double(word[2:], count)
        return find_three_double(word[1:])
    return True

def main():
    three_double = []

    for line in open('words.txt'):
        if find_three_double(line.strip()):
            three_double.append(line.strip())
    print(len(three_double))
    print(three_double)
    
main()
```

```python
"""This module contains a code example related to

Think Python, 2nd Edition
by Allen Downey
http://thinkpython2.com

Copyright 2015 Allen Downey

License: http://creativecommons.org/licenses/by/4.0/
"""

from __future__ import print_function, division


def is_triple_double(word):
    """Tests if a word contains three consecutive double letters.
    
    word: string

    returns: bool
    """
    i = 0
    count = 0
    while i < len(word)-1:
        if word[i] == word[i+1]:
            count = count + 1
            if count == 3:
                return True
            i = i + 2
        else:
            i = i + 1 - 2*count
            count = 0
    return False


def find_triple_double():
    """Reads a word list and prints words with triple double letters."""
    fin = open('words.txt')
    for line in fin:
        word = line.strip()
        if is_triple_double(word):
            print(word)


print('Here are all the words in the list that have')
print('three consecutive double letters.')
find_triple_double()
print('')
```

```python
## 9.7 找到num个连续双字符,用户在主函数中输入num
## 注意函数中的参数顺序问题

def find_num_double(word, num=0, count=0):
    if count < num:
        if len(word) < 2:
            return False
        if word[0] == word[1]:
            count += 1
            return find_num_double(word[2:], num, count)
        return find_num_double(word[1:], num)
    return True

def main():
    num_double = []
    for line in open('words.txt'):
        if find_num_double(line.strip(), 3):
            num_double.append(line.strip())
    print(len(num_double))
    print(num_double[2])
    
main()
```

```python
## 9.8

def reverse(string):
    i = 0
    j = len(string) - 1
    while i < j:
        if string[i] != string[j]:
            return False
        i = i + 1
        j = j - 1
    return True


def find_number(number):
    if reverse(str(number)[2:]):
        if reverse(str(number+1)[1:]):
            if reverse(str(number+2)[1:5]):
                if reverse(str(number+3)):
                    return True
    return False
    
def main():
#     want_number = 0
    want_list = []
    for i in range(100000, 999999):
        if find_number(i):
            want_list.append(i)
#             want_number = i
#             break     ## 找到一个就可以,退出循环
#     print(want_number)
    print(len(want_list))
    print(want_list)
    
main()
```

```python
"""This module contains a code example related to

Think Python, 2nd Edition
by Allen Downey
http://thinkpython2.com

Copyright 2015 Allen Downey

License: http://creativecommons.org/licenses/by/4.0/
"""

from __future__ import print_function, division


def str_fill(i, n):
    """Returns i as a string with at least n digits.

    i: int
    n: int length

    returns: string
    """
    return str(i).zfill(n)


def are_reversed(i, j):
    """Checks if i and j are the reverse of each other.

    i: int
    j: int

    returns:bool
    """
    return str_fill(i, 2) == str_fill(j, 2)[::-1]


def num_instances(diff, flag=False):
    """Counts the number of palindromic ages.

    Returns the number of times the mother and daughter have
    palindromic ages in their lives, given the difference in age.

    diff: int difference in ages
    flag: bool, if True, prints the details
    """
    daughter = 0
    count = 0
    while True:
        mother = daughter + diff

        # assuming that mother and daughter don't have the same birthday,
        # they have two chances per year to have palindromic ages.
        if are_reversed(daughter, mother) or are_reversed(daughter, mother+1):
            count = count + 1
            if flag:
                print(daughter, mother)
        if mother > 120:
            break
        daughter = daughter + 1
    return count
    

def check_diffs():
    """Finds age differences that satisfy the problem.

    Enumerates the possible differences in age between mother
    and daughter, and for each difference, counts the number of times
    over their lives they will have ages that are the reverse of
    each other.
    """
    diff = 10
    while diff < 70:
        n = num_instances(diff)
        if n > 0:
            print(diff, n)
        diff = diff + 1

print('diff  #instances')
check_diffs()

print()
print('daughter  mother')
num_instances(18, True)
```
