```python
## 9.4

def uses_only(word, available):
    for letter in word:
        if letter not in available:
            return False
    return True
    
print(uses_only('ts', 'tus'), uses_only('ta','tus'))
```

```python
## 9.6

def is_abecedarian(word):
    first = word[0]
    for i in range(1, len(word)):
        if ord(word[i]) < ord(first):  # 直接比较 word[i]  first
            return False
        first = word[i]
    return True

def main():
    count = 0
    for line in open('words.txt'):
        if is_abecedarian(line.strip()):
#             print(line.strip())
            count += 1
    print(count)
    
    
main()
```

```python
## 9.6 利用for循环

def is_abecedarian(word):
    previous = word[0]
    for c in word:
        if c < previous:
            return False
        previous = c
    return True

def main():
    count = 0
    for line in open('words.txt'):
        if is_abecedarian(line.strip()):
            count += 1
    print(count)
    
main()
```

```python
## 采用递归

def is_abecedarian(word):
    if len(word) <= 1:
        return True
    if word[0] > word[1]:
        return False
    return is_abecedarian(word[1:])

def main():
    count = 0
    for line in open('words.txt'):
        if is_abecedarian(line.strip()):
            count += 1
    print(count)
        
main()
```

```python
## 9.7 找到一对连续双字符

def find_one_double(word):
    
    if len(word) < 2:
        return False
    if word[0] == word[1]:
        
        return True
    return find_one_double(word[1:])

def main():
    one_double = []
    for line in open('words.txt'):
        if find_one_double(line.strip()):
            one_double.append(line.strip())
    print(len(one_double))
    
main()
```

```python
## 9.7 找到三个连续双字符

def find_three_double(word, count=0):
    if count != 3:
        if len(word) < 2:
            return False
        if word[0] == word[1]:
            count += 1
            return find_three_double(word[2:], count)
        return find_three_double(word[1:])
    return True

def main():
    three_double = []

    for line in open('words.txt'):
        if find_three_double(line.strip()):
            three_double.append(line.strip())
    print(len(three_double))
    print(three_double[0])
    
main()
```
