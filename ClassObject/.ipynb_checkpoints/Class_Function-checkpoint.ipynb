```python
## 16.1 时间

class Time:
    """Represents the time of day.
    arributes: hour, minute, second
    """

time = Time()
time.hour = 10
time.minute = 41
time.second = 29
```

```python
def print_time(tm):
    print('%.2d:%.2d:%.2d' % (tm.hour, tm.minute, tm.second))
    
print_time(time)
```

```python
## t1 is after t2?
## t1 and t2 both in the one day and 24h

import copy

def is_after(t1, t2):
    time1 = t1.hour*3600 + t1.minute*60 + t1.second
    time2 = t2.hour*3600 + t2.minute*60 + t2.second
    
    return time1 > time2

t1 = copy.copy(time)
t2 = copy.copy(time)
t2.hour = 9
t2.second = 59

print(is_after(t1, t2))
```

```python
## 16.2 纯函数

def add_time(t1, t2):
    sum = Time()
    sum.hour = t1.hour + t2.hour
    sum.minute = t1.minute + t2.minute 
    sum.second = t1.second + t2.second
    return sum

def main():
    start = Time()
    start.hour = 9
    start.minute = 45
    start.second = 0
    
    duration = Time()
    duration.hour = 1
    duration.minute = 35
    duration.second = 0
    
    done = add_time(start, duration)
    print_time(done)
    
main()
```

```python
## 16.2 改进纯函数

def add_time_modified(t1, t2):
    sum = Time()
    sum.hour = t1.hour + t2.hour
    sum.minute = t1.minute + t2.minute 
    sum.second = t1.second + t2.second
    
    if sum.second >= 60:
        sum.second -= 60
        sum.minute += 1
        
    if sum.minute >= 60:
        sum.minute -= 60
        sum.hour += 1
        
    return sum

def main():
    start = Time()
    start.hour = 9
    start.minute = 45
    start.second = 0
    
    duration = Time()
    duration.hour = 1
    duration.minute = 35
    duration.second = 0
    
    done = add_time_modified(start, duration)
    print_time(done)
    
main()
```

```python
## 重写 add_time

def time_to_int(time):
    seconds = time.hour * 3600 + time.minute * 60 + time.second
    return seconds

def int_to_time(seconds):
    time = Time()
    minutes, time.second = divmod(seconds, 60)
    time.hour, time.minute = divmod(minutes, 60)
    return time

def add_time(t1, t2):
    seconds = time_to_int(t1) + time_to_int(t2)
    return int_to_time(seconds)

def main():
    
    start = Time()
    start.hour = 9
    start.minute = 45
    start.second = 0
    
    duration = Time()
    duration.hour = 1
    duration.minute = 35
    duration.second = 0

    print_time(add_time(start, duration))
    
main()
```

```python
## 16.5 调试, 检测不变式

def time_to_int(time):
    seconds = time.hour * 3600 + time.minute * 60 + time.second
    return seconds

def int_to_time(seconds):
    time = Time()
    minutes, time.second = divmod(seconds, 60)
    time.hour, time.minute = divmod(minutes, 60)
    return time

def valid_time(time):
    """ 编写代码检查不变式帮助检测错误"""
    if time.hour < 0 or time.minute < 0 or time.second < 0:
        return False
    if time.minute >= 60 or time.second > 60:
        return False
    return True

def add_time(t1, t2):
    if not valid_time(t1) or not valid_time(t2):
        raise ValueError('invalid Time object in add time.')
    seconds = time_to_int(t1) + time_to_int(t2)
    return int_to_time(seconds)

def main():
    
    start = Time()
    start.hour = 9
    start.minute = 65
    start.second = 0
    
    duration = Time()
    duration.hour = 1
    duration.minute = 35
    duration.second = 0

    print_time(add_time(start, duration))
    
main()
```

```python
## 16.5 调试, 检测不变式

def time_to_int(time):
    seconds = time.hour * 3600 + time.minute * 60 + time.second
    return seconds

def int_to_time(seconds):
    time = Time()
    minutes, time.second = divmod(seconds, 60)
    time.hour, time.minute = divmod(minutes, 60)
    return time

def valid_time(time):
    """ 编写代码检查不变式帮助检测错误"""
    if time.hour < 0 or time.minute < 0 or time.second < 0:
        return False
    if time.minute >= 60 or time.second > 60:
        return False
    return True

def add_time(t1, t2):
    """assert 语句可以区分处理普通条件的代码和检查错误代码，在失败的条件下抛出异常"""
    assert valid_time(t1) and valid_time(t2)
    seconds = time_to_int(t1) + time_to_int(t2)
    return int_to_time(seconds)

def main():
    
    start = Time()
    start.hour = 9
    start.minute = 65
    start.second = 0
    
    duration = Time()
    duration.hour = 1
    duration.minute = 35
    duration.second = 0

    print_time(add_time(start, duration))
    
main()
```

```python
## 16.7 练习__Exercise 16.1


import copy


def print_time(t):
    print("%.2d:%.2d:%.2d" % (t.hour, t.minute, t.second))
    

def valid_time(t):
    if t.hour < 0 or t.minute < 0 or t.second < 0:
        return False
    if t. hour > 24 or t.minute >= 60 or t.second >= 60:
        return False
    return True


def int_to_time(seconds):
    t_next = Time()
    t_next.minute, t_next.second = divmod(seconds, 60)
    t_next.hour, t_next.minute = divmod(t_next.minute, 60)
    t_next.hour = t_next.hour % 24
    return t_next

    
def time_to_int(t):
    seconds = t.hour * 3600 + t.minute * 60 + t.second
    return seconds


def mul_time(t, num):
    if not valid_time(t):
        raise ValueError("valid time object in mul_time.")
    mul_t = int_to_time(time_to_int(t) * num)
    if not valid_time(mul_t):
        raise ValueError("valid time pbject after mul the time.")
    return mul_t


def main():
    
    te = copy.copy(time)
    number = 3
    
    print_time(mul_time(te, number))

main()
```
